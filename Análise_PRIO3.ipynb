Vamos calcular qual foi a maior queda em um ano da PRIO3 e sua maior ascensão

```python
import pandas as pd
from pandas_datareader import data as pdr
import numpy as np
import yfinance as yf
```

```python
cot = yf.download("PRIO3.SA", "2010-10-25", "2023-06-30")["Adj Close"]
cot
```

```python
cotacoes_maxdd = cot.to_frame()
cotacoes_maxdd['ano'] = cotacoes_maxdd.index.year


cotacoes_maxdd['maxima_do_ano'] = cotacoes_maxdd.groupby("ano")["Adj Close"].cummax()

cotacoes_maxdd
```

```python
cotacoes_maxdd["Quedas"] = cotacoes_maxdd["Adj Close"]/cotacoes_maxdd["maxima_do_ano"] - 1

cotacoes_maxdd
```

```python
cotacoes_maxdd.groupby("ano")["Quedas"].min()
```

```python
# Nessa Análise podemos perceber que o comprador que comprou mais no topo e vendeu mais na queda perdeu entre
# 8% a 78% a depender do ano
```

```python
cotacoes_maxdd['Minima_do_ano'] = cotacoes_maxdd.groupby("ano")["Adj Close"].cummin()
cotacoes_maxdd
```

```python
cotacoes_maxdd["Altas"] = cotacoes_maxdd["Adj Close"]/cotacoes_maxdd["Minima_do_ano"] -1

cotacoes_maxdd
```

```python
cotacoes_maxdd.groupby("ano")["Altas"].max()
```

```python
# Tendo em vista esses dados, nós podemos perceber que o ano de 2020 um investidor
# comprando no melhor momento ele consegue gerar um resultado de até 5x o investido;
```
